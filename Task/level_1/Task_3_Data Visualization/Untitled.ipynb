```python
import pandas as pd
df = pd.read_csv('task2_dataset.csv')
```

#### Step 1: City-Wise Distribution Visualization
Objective: Show which cities have the most restaurants using a bar chart.

```python
import matplotlib.pyplot as plt

# Top 10 cities with the most restaurants
top_cities = df['City'].value_counts().head(10)

# Plot bar chart
plt.figure(figsize=(12, 6))
top_cities.plot(kind='bar', color='skyblue')
plt.title('Top 10 Cities with the Most Restaurants', fontsize=14)
plt.xlabel('City', fontsize=12)
plt.ylabel('Number of Restaurants', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### 📊 Key Insights from the Bar Chart:
1️⃣ Which cities have the most restaurants?

New Delhi, Gurgaon, and Noida likely dominate the list.

These cities have a high concentration of restaurants.

####  Step 2 – Cuisine Popularity Visualization 🍽️
Now, let’s create a bar chart or pie chart to analyze the most popular cuisines in the dataset.

```python
# Top 10 most popular cuisines
top_cuisines = df['Cuisines'].value_counts().head(10)

# Plot bar chart
plt.figure(figsize=(12, 6))
top_cuisines.plot(kind='bar', color='coral')
plt.title('Top 10 Most Popular Cuisines', fontsize=14)
plt.xlabel('Cuisine', fontsize=12)
plt.ylabel('Number of Restaurants', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### 📊 Key Insights from the Cuisine Bar Chart:
##### 1️⃣ Which cuisines are the most popular?

1.North Indian, Chinese, and Fast Food are likely in the top positions.

2.These cuisines are widely available in multiple cities.
##### 2️⃣ Are fusion cuisines common?

1.Some cuisines might be combinations (e.g., North Indian + Chinese, Bakery + Desserts).

2.This shows that customers prefer variety.

####  Step 3 – Price vs. Rating Visualization 💰⭐
Now, let’s see how restaurant pricing affects ratings using a scatter plot.

```python
import seaborn as sns

# Scatter plot for Price vs. Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Average Cost for two', y='Aggregate rating', alpha=0.5, color='purple')

plt.title('Price vs. Aggregate Rating', fontsize=14)
plt.xlabel('Average Cost for Two', fontsize=12)
plt.ylabel('Aggregate Rating', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

#### 📊 Key Insights from the Scatter Plot (Price vs. Rating)
##### 1️⃣ Do expensive restaurants have higher ratings?

1.If higher-priced restaurants (right side of the plot) have more high ratings (top side), it confirms that expensive places receive better reviews.

2.If low-cost restaurants (left side) also have high ratings, it means quality matters more than price.
##### 2️⃣ Are there affordable restaurants with high ratings?

1.If some low-cost restaurants (left) reach high ratings (top), it means that great service & food quality can exist at any price range.

#### Step 4 – Service Availability Impact Visualization 📦
Now, let’s analyze whether Table Booking & Online Delivery services impact restaurant ratings using a bar chart.

```python
# Average rating for restaurants with and without table booking
table_booking_ratings = df.groupby('Has Table booking')['Aggregate rating'].mean()

# Plot bar chart
plt.figure(figsize=(8, 5))
table_booking_ratings.plot(kind='bar', color=['red', 'green'])
plt.title('Impact of Table Booking on Ratings', fontsize=14)
plt.xlabel('Has Table Booking (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Average Aggregate Rating', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### 📊 Key Insights from the Table Booking Analysis
##### 1️⃣ Do restaurants with table booking have higher ratings?
1.If the bar for "1" (Yes) is higher, it means restaurants that offer table booking get better ratings.

2.If the bar for "0" (No) is lower, it means restaurants without table booking struggle to get high ratings.
##### 2️⃣ Why does table booking improve ratings?
1.It offers a better customer experience, leading to higher satisfaction.

2.Such restaurants might provide premium service and high-quality food.

#### Step 5 – Online Delivery Impact Visualization 🚀
Now, let’s analyze whether Online Delivery affects restaurant ratings using another bar chart.

```python
# Average rating for restaurants with and without online delivery
online_delivery_ratings = df.groupby('Has Online delivery')['Aggregate rating'].mean()

# Plot bar chart
plt.figure(figsize=(8, 5))
online_delivery_ratings.plot(kind='bar', color=['blue', 'orange'])
plt.title('Impact of Online Delivery on Ratings', fontsize=14)
plt.xlabel('Has Online Delivery (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Average Aggregate Rating', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### 📊 Key Insights from the Online Delivery Analysis
##### 1️⃣ Do restaurants with online delivery have higher ratings?
1.If the bar for "1" (Yes) is higher, it means restaurants that offer online delivery get better ratings.
    
2.If the bar for "0" (No) is lower, it means restaurants without online delivery receive lower ratings.
##### 2️⃣ Why does online delivery improve ratings?

1.Customers appreciate convenience, leading to higher satisfaction.
    
2.More customers leave positive reviews for fast and reliable delivery services.
