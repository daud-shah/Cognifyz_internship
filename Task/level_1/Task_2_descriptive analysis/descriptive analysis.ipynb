> Compare Ratings of Restaurants with vs. without These Services

```python
# Average rating for restaurants with and without table booking
print(df.groupby('Has Table booking')['Aggregate rating'].mean())

# Average rating for restaurants with and without online delivery
print(df.groupby('Has Online delivery')['Aggregate rating'].mean())
```

📊 Key Insights from Service Availability Analysis

🏷️ Impact of Table Booking on Ratings:

Restaurants without Table Booking (0): 2.56 ⭐ (Lower rating)

Restaurants with Table Booking (1): 3.44 ⭐ (Higher rating)

📌 What This Means:

✅ Restaurants with table booking get better ratings because:

Customers may prefer a more organized, hassle-free experience.
    
Likely to be higher-end restaurants with better service.
    
📦 Impact of Online Delivery on Ratings:
    
Restaurants without Online Delivery (0): 2.47 ⭐ (Lower rating)
    
Restaurants with Online Delivery (1): 3.25 ⭐ (Higher rating)

📌 What This Means:

✅ Restaurants with online delivery get higher ratings because:

Customers value convenience and easy access to food.
    
More people leave reviews due to a higher order volume.
