```python
import pandas as pd 
df = pd.read_csv('level_2_dataset.csv')
```

### Step 1: Visualizing the Distribution of Ratings
In this step, we will analyze how restaurant ratings are distributed using visualizations like histograms and bar plots.

#### Visualization 1: Histogram of Ratings

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style("whitegrid")

# Create histogram
plt.figure(figsize=(8, 5))
sns.histplot(df['Aggregate rating'], bins=10, kde=True, color="blue")

# Add labels and title
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Restaurants")
plt.title("Distribution of Restaurant Ratings")

# Show the plot
plt.show()
```

 #### Visualization 2: Bar Plot of Rating Counts

```python
# Count of each unique rating value
rating_counts = df['Aggregate rating'].value_counts().sort_index()

# Create bar plot
plt.figure(figsize=(8, 5))
sns.barplot(x=rating_counts.index, y=rating_counts.values, palette="coolwarm")

# Add labels and title
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Restaurants")
plt.title("Count of Each Rating Value")

# Show the plot
plt.show()
```

### Visualization 3: Violin Plot of Ratings

```python
# Create a violin plot for restaurant ratings
plt.figure(figsize=(8, 5))
sns.violinplot(x=df['Aggregate rating'], color="purple")

# Add title and labels
plt.title("Violin Plot of Restaurant Ratings")
plt.xlabel("Aggregate Rating")

# Show the plot
plt.show()
```

#### Visualization 4: Box Plot of Ratings

```python
# Create a box plot for restaurant ratings
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Aggregate rating'], color="lightblue")

# Add title and labels
plt.title("Box Plot of Restaurant Ratings")
plt.xlabel("Aggregate Rating")

# Show the plot
plt.show()
```

#### visualization 5: Pie Chart of Rating Distribution

```python
# Group rating counts
rating_distribution = df['Aggregate rating'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(rating_distribution, labels=rating_distribution.index, autopct='%1.1f%%', colors=sns.color_palette("coolwarm", len(rating_distribution)))

# Add title
plt.title("Percentage of Restaurants in Each Rating Category")

# Show the plot
plt.show()
```

### Step 2: Compare Ratings Across Cuisines and Cities
Now, let‚Äôs check how different cuisines and cities compare in terms of ratings. 

####  Visualization 6: Average Rating by Cuisine:

```python
# Calculate the average rating for each cuisine
cuisine_ratings = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

# Create a bar plot
plt.figure(figsize=(10, 5))
sns.barplot(x=cuisine_ratings.values, y=cuisine_ratings.index, palette="viridis")

# Add title and labels
plt.xlabel("Average Rating")
plt.ylabel("Cuisine Type")
plt.title("Top 10 Highest Rated Cuisines")

# Show the plot
plt.show()
```

#### Visualization 7: Average Rating by City
This will show which cities have the highest-rated restaurants on average.

```python
# Calculate the average rating for each city
city_ratings = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

# Create a bar plot
plt.figure(figsize=(10, 5))
sns.barplot(x=city_ratings.values, y=city_ratings.index, palette="mako")

# Add title and labels
plt.xlabel("Average Rating")
plt.ylabel("City")
plt.title("Top 10 Cities with Highest Rated Restaurants")

# Show the plot
plt.show()
```

### Step 3: Visualizing Feature Relationships
Now, let‚Äôs explore how different features relate to restaurant ratings.

#### Visualization 8: Price Range vs. Aggregate Rating
This will help analyze whether expensive restaurants receive higher ratings.

```python
# Create a box plot for price range vs. ratings
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Price range'], y=df['Aggregate rating'], palette="coolwarm")

# Add title and labels
plt.xlabel("Price Range")
plt.ylabel("Aggregate Rating")
plt.title("Price Range vs. Restaurant Ratings")

# Show the plot
plt.show()
```

####  Visualization 9: Votes vs. Aggregate Rating
This scatter plot will show if restaurants with more votes tend to have higher ratings.

```python
# Create a scatter plot for votes vs. ratings
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Votes'], y=df['Aggregate rating'], alpha=0.5, color="blue")

# Add title and labels
plt.xlabel("Number of Votes")
plt.ylabel("Aggregate Rating")
plt.title("Votes vs. Restaurant Ratings")

# Show the plot
plt.show()
```

#### Visualization 10: Table Booking vs. Ratings
This will show if restaurants that offer table booking receive higher ratings.

```python
# Create a bar plot for table booking vs. ratings
plt.figure(figsize=(8, 5))
sns.barplot(x=df['Has Table booking'], y=df['Aggregate rating'], palette="magma")

# Add title and labels
plt.xlabel("Has Table Booking (0 = No, 1 = Yes)")
plt.ylabel("Average Rating")
plt.title("Impact of Table Booking on Restaurant Ratings")

# Show the plot
plt.show()
```
